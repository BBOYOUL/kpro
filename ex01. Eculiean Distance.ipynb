# Eculidean Distance
베이직한 두 점사이의 거리를 측정하는 알고리즘 

```python
import numpy as np
import pandas as pd


# (height, weight, age, sex(man=0, woman=1))
woman1 = pd.Series([160, 50, 23, 1], name='박미영')
woman2 = pd.Series([155, 48, 27, 1], name='아이윳')
man1 = pd.Series([175, 72, 31, 0], name='이순신')
man2 = pd.Series([176, 75, 28, 0], name='강감찬')

lst = [woman1, woman2, man1, man2]
lst
```

- 유클리디언 거리 계산 함수

```python

def euclidean_dist(x, y):
    return np.sqrt(np.sum((x - y) ** 2))
```

```python
def recommend_by_dist(li):
    for i in range(0, len(lst)-1):
        for j in range(i+1, len(lst)):
            x, y = lst[i], lst[j]
            print('distance between {} and {}: {}'.format(x.name, y.name, euclidean_dist(x, y)))

recommend_by_dist(lst)
```
